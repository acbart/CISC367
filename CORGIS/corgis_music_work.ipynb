```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\megan\\Desktop\\CISC367\\Week 5\\music_data\\data.csv")
df.head()
```

```python
# Gets the first listed artist from the artists list and puts it into its own column for simplicity's sake
def primary_artist(artists: str) -> str:
    artists = artists.split("', '")
    if len(artists) > 1:
        return artists[0][2:]
    else:
        return artists[0][2:-2]
df['artist'] = df['artists'].map(primary_artist)
df
```

```python
# Putting any additional artists into a list in their own column; new column has an empty list if there are no additional artists
def other_artists(artists: str) -> [str]:
    split_artists = artists.split(",")
    others = []
    i = 1
    while i < len(split_artists):
        if i == len(split_artists) - 1:
            others.append(split_artists[i][2:-2])
        else:
            others.append(split_artists[i][2:-1])
        i += 1
    return others

df['additional_artists'] = df['artists'].map(other_artists)
df[df['additional_artists'].apply(lambda x: len(x)) > 0].head()
```

```python
# Shows empty list value for songs with no additional artists
df.head()
```

```python
# Dropping clunky artists column 
df = df.drop(columns = ['artists'])
df.head()
```

```python
# Splitting date field into month and day fields, have values -1 if exact release date not given
def get_month(date: str) -> int:
    split_date = date.split("-")
    month = -1
    if len(split_date) > 1:
        month = int(split_date[1])
    return month

def get_day(date: str) -> int:
    split_date = date.split("-")
    day = -1
    if len(split_date) > 2:
        day = int(split_date[2])
    return day

df['month'] = df['release_date'].apply(get_month)
df['day'] = df['release_date'].apply(get_day)
df.head()
```

```python
# Remove clunky release date column
df = df.drop(columns = 'release_date')
```

```python
# Bringing the data down from 32mb to 27mb but dropping unique song IDs that share both the artist and title (i.e. different versions of the same song)
df = df.drop_duplicates(subset = ['artist', 'name'])
```

```python
# Rearranging headers to be more logical (from things you observe to things you calculate)
df = df[['name', 'artist', 'additional_artists', 'id', 'year', 'month', 'day', 'duration_ms', 'explicit', 'popularity', 'key', 'mode', 'tempo', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'valence']]
```
